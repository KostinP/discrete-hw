1. Пусть число $A=\dfrac{3}{14}$, в качестве его приближения взято число $a=0.22$. Найдите абсолютную и относительную погрешности.

**Решение.** $\dfrac{3}{14} = 0.21428571428571427...$

Абсолютная погрешность: $\Delta = |0.21428571428571427 - 0.22| = 0.005714285714285727$

Относительная погрешность: $\delta_x = \dfrac{0.005714285714285727}{0.21428571428571427} = 0.0266666666666667$
